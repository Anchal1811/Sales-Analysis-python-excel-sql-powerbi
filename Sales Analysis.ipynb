```python
import pandas as pd

# The 'r' ensures the backslashes are read correctly as a file path
file_path = r"C:\Users\DELL\Desktop\sales_raw_500.xlsx"
df = pd.read_excel(file_path)

# Show the first few rows to confirm it worked
df.head()
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

```python
df.dropna(subset=['Order_Date'])
```

```python
df['Profit'] = df['Sales']-df['Cost']
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

```python
# This creates 'sales_cleaned.csv' in your working folder
df.to_csv('sales_cleaned.csv', index=False)
print("File saved successfully!")
```
